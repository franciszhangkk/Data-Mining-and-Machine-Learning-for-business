```python
from sklearn import preprocessing
import numpy as np

#数据预处理
X = df_loan[['annual_inc','int_rate']].values
X = preprocessing.scale(X)
y = df_loan[['debt_settlement_flag']].values
```

```python
def sigmoid_function(X):
    return np.exp(X) / (1 + np.exp(X))
```

```python
def myLogisticGD (X,y,alpha,numIter):
    beta = np.array([[0],[0],[0]])
    X = np.insert(X, 0, values=1, axis=1)
    XT = X.transpose()
    N = X.shape[0]
    for i in range(0, numIter):
        predictions = sigmoid_function(np.dot(X, beta))
        diff = predictions - y
        gradient = np.dot(XT,diff)/N
        print("Iteration: {0} | Gradient: {1}".format(i,gradient.reshape((1,3))))
        beta = beta - alpha*gradient
    return beta
        
    
```

```python
beta = myLogisticGD(X,y,1,1000)
```

```python
def compute_loss(X,y,beta1):
    N = X.shape[0]
    X = np.insert(X, 0, values=1, axis=1)
    get_sum = sum(y*np.log(sigmoid_function(np.dot(X, beta1)))+(1-y)*np.log(1-sigmoid_function(np.dot(X, beta1))))
    loss = -get_sum/N
    return loss
```
